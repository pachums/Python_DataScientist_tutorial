# Programación Orientada a Objetos (POO)

La programación dirigida a objetos consiste en crear unas 'plantillas' quee specifican los atributos y las interacciones que los objetos que siguen esa plantilla tendrán.

Una vez creados los objetos pueden interactuar para llevar a cabo una función específica.



La programación dirigida a objetos tiene ciertas ventajas:
**Ventajas:**

 - Código más facil de reutilizar en otros proyectos
 
 - Más facil añadir cambios, actualizar programas
 
 - Más intuitivo a la hora de la aplicación
 
 - Más comprensible, al igual que las funciones 'en lineas generales'



La programación dirigida a objetos tiene **4 conceptos básicos**. Estos son

- Encapsulación
    
    
      Añadir propiedades no visibles (privadas) y acceder a ellas mediante métodos
      
      > No podemos acceder a las cuerdas del piano y piezas del piano, tan solo tocar sus teclas.
- Abstracción


      Para conseguir una funcionalidad/operación que relaciona elementos, puede ser tan sencillo como llamar a
        una función que realiza múltiples operaciones.
    
      > Al igual que cuando utilizamos el mando a distancia de la tele (tenemos botones diversos y relaciones entre estos y el menú)
    
- Herencia
    
      Propiedades que se toman por defecto. 
      
      No se construye de cero. Al igual que un caballo es un animal, y por tanto comparte cierta naturaleza con otros animales, podemos describir que es un animal y después tener en cuenta sus propiedades previamente establecidas.

- Polimorfismo
    
      Capacidad de comportarse de varias formas

## Estructura de clases - primeros ejemplos

Las clases se componen de 
 - **Atributos** (datos) valores asignados a variables embebidas en la clase
 - **Métodos** (funciones) que representan la interacción con esos atributos.

Veamos cómo se escribe una clase en python...

```python
# Declaración de clase
class Pokemon(object):
    
    edicion = 'rojo_fuego'
    def __init__(self, nombre_pokemon, tipo_pokemon, tamaño):
        
        # ATRIBUTOS
        # > Nombre Pokemon
        self.nombre = nombre_pokemon
        # > Tipo Pokemon
        self.tipo = tipo_pokemon
        # > Tamaño del Pokemon
        self.tamaño = tamaño
        
        # > Nivel
        self.nivel = np.random.randint(3,7) 
        
```

En este ejemplo hay un método, típico en las clases, que es el '\_\_init\_\_()' (constructor), que inicializa los atributos de la clase cuando una instancia de esta se crea.

Este no es necesario, pero se recomienda puesto que las variables fuera del '__init__()' tienen el peligro de ser accedidas antes de ser definidas.

Se pueden definir funciones dentro del espacio (namespace) de la clase, estas son consideradas **métodos**

```python
class Pokemon_v2(object): 
    
    edicion = 'rojo_fuego'
    def __init__(self, nombre_pokemon, tipo_pokemon, tamaño):
        
        # ATRIBUTOS
        # > Nombre Pokemon
        self.nombre = nombre_pokemon
        # > Tipo Pokemon
        self.tipo = tipo_pokemon
        # > Tamaño del Pokemon
        self.tamaño = tamaño
        
        # > Nivel
        self.__nivel = np.random.randint(3,7) 
        
    def subir_nivel(self):
        # Subimos nivel si es posible
        if self.__nivel < 100:
            print(f'{self.nombre} ha subido del Nº{self.__nivel} al Nº{self.__nivel+1}.')
            self.__nivel += 1
        else:
            print(f'{self.nombre} no puede subir de nivel.')
```

El **método** 'subir\_nivel()' no utiliza ninguna variable de entrada, pero hemos introducido la palabra **'self'**:

    Ciertamente, todos los métodos, includio el *__init__()*, requieren pasar el objeto en cuestión como primer parámetro. Esto se hace por convención utilizando la palabra *self*.
    
    Veremos que hay 2 excepciones

**Métodos mágicos**

 Hay métodos que se dan en la forma '\_\_*nombre\_de\_método*\_\_'  (como el init). Estos son especiales en Python.
 Por ejemplo, los operadores de una clase. Estos comparan dos instancias. [Más sobre métodos mágicos](https://towardsdatascience.com/magic-methods-in-python-by-example-16b6826cae5c)

```python
class Pokemon_v3(object): 
    
    edicion = 'rojo_fuego'
    def __init__(self, nombre_pokemon, tipo_pokemon, tamaño):
        
        # ATRIBUTOS
        # > Nombre Pokemon
        self.nombre = nombre_pokemon
        # > Tipo Pokemon
        self.tipo = tipo_pokemon
        # > Tamaño del Pokemon
        self.tamaño = tamaño
        
        # > Nivel
        self.nivel = np.random.randint(3,7) 
        
    def subir_nivel(self):
        # Subimos nivel si es posible
        if self.nivel < 100:
            print(f'{self.nombre} ha subido del Nº{self.nivel} al Nº{self.nivel+1}.\n')
            self.nivel += 1
        else:
            print(f'{self.nombre} no puede subir de nivel.\n')
    
    # Incluimos OPERADORES que usen el atributo 'nivel' como referencia
    def __gt__(self, other):
        return self.nivel > other.nivel 
    def __lt__(self, other):
        return self.nivel < other.nivel
    def __ge__(self, other):
        return self.nivel >= other.nivel
    def __le__(self, other):
        return self.nivel <= other.nivel # Recordad que el return no requiere de paréntesis
    
    # Incluimos un 'método mágico' de imprimir
    def __str__(self):
        # Imprimimos atributos
        return(f'{self.nombre} (Nº_{self.nivel} | {self.tipo} | {self.tamaño})')
```

## Herencia (única o múltiple)

Monkey patching

Ahora cambiamos de Pokémons a **PokéBalls**, para ver el ejemplo de la **herencía de clases**.

```python
class Ball:
    
    def __init__(self, probabilidad, umbral_nivel):
        self.__p = probabilidad              # Probabilidad de capturar pokemon
        self.__umbral = umbral_nivel         #  si su nivel está por debajo del umbral
    
    def lanzar(self, pokemon):
        
        # Comprobar si el Pokemon está dentro de la posibilidad de la 'Ball()'
        if pokemon.nivel > self.__umbral:
            print(f'{pokemon.nombre} no ha sido capturado.')
            return(False)
        
        # Jugar a probabilidades
        else:
            # Captura con probabilidad 'self.__p'
            r = np.random.random_sample() # Numero aleatorio de 0 a 1
            if r < self.__p:
                print(f'¡Enhorabuena! {pokemon.nombre} ha sido capturado.')
                return(True)
            else:
                print(f'Lastima, {pokemon.nombre} no ha sido capturado.')
                return(False)

class PokeBall(Ball):
    def __init__(self):
        super().__init__(0.3, 21)

class SuperBall(Ball):
    def __init__(self):
        super().__init__(0.5, 34)
        
class UltraBall(Ball):
    def __init__(self):
        super().__init__(0.6, 55)
        
class MasterBall(Ball):
    def __init__(self):
        super().__init__(1.0, 100)
```

Se dice que la clase 'Ball()' es la *BaseClass* (Padre/Madre) y las otras son las *DerivedClass* (Hijo/Hija).

Esta última tiene los atributos y métodos por **herencia**.

```python
# Para comprobar herencia
print(issubclass(SuperBall, Ball))
print(issubclass(SuperBall, object))
print(issubclass(Ball, SuperBall))
```

```python
pb = PokeBall()
sb = SuperBall()
ub = UltraBall()
mb = MasterBall()
```

```python
print(isinstance(pb, Ball))
print(isinstance(ub, object))
print(isinstance(pb, PokeBall))
print(isinstance(pb, UltraBall))
print(isinstance(pb, SuperBall))
```

## Ejemplo: Sala de cine

```python
class Asiento:
    '''Clase que simula un asiento con atributo de disponibilidad y una posicion fija (tupla) \
      para el asiento.     
    '''
    
    def __init__(self, posicion_en_sala):

        self.__disponible = True
        assert(isinstance(posicion_en_sala, tuple))
        self.__posicion = posicion_en_sala
        
    def ocupar(self):
        ''' Ocupar el asiento '''
        
        if self.__disponible:
            self.__disponible = False
            return(True) # Exitoso
        else:
            return(False)
        
    def liberar(self):
        ''' Liberar el asiento '''
        
        if not self.__disponible:
            self.__disponible = True
            return(True) # Exitoso
        else:
            return(False)   

    # GETTERS
    def esta_disponible(self):
        return(self.__disponible)
    
    def posicion(self):
        return(self.__posicion)
    
    # Para imprimirlo
    def __str__(self):
        # Imprimimos atributos
        disponible_str = 'disponible' if self.__disponible else 'ocupado' 
        return(f'Posicion {self.__posicion} - {disponible_str}')

    
class Sala:
    'Clase para añadir instancias de Asiento() en posiciones fijas (tuplas) dentro de un iterable \
      **posiciones_asientos**'
    
    def __init__(self, posiciones_asientos):
        
        self.n_asientos = len(posiciones_asientos)
        
        self.__asientos_dict = {}
        for posicion in posiciones_asientos:
            # Creamos un asiento en la posicion introducida
            self.__asientos_dict[posicion] = Asiento(posicion)
        
    
    def ocupar_asientos(self, *posiciones_a_ocupar):
        
        for posicion in posiciones_a_ocupar:
            # Comprobamos que la posicion del asiento existe
            assert(posicion in self.__asientos_dict)
            
            # Ocupamos el asiento y mostamos mensaje dependiendo del éxito de la
            #  transacción
            if self.__asientos_dict[posicion].ocupar():
                print(f'Asiento {posicion} correctamente adquirido.')
            else:
                print(f'Asiento {posicion} está ocupado.')
    
    # SHOWER
    def asientos_disponibles(self):
        for asiento in self.__asientos_dict.values():
            if asiento.esta_disponible():
                print(asiento)
        
    # GETTER
    def asientos_dict(self):
        return self.__asientos_dict
    
    
class Sala_cine(Sala):
    ''' 
    Simula una sala de cine mediante instancias de Asiento() en todas las combinaciones de posiciones de fila, **posiciones_filas** \
          y **posiciones_columnas**, excluyendo las posiciones en la variable opcional **huecos**.
          
    Incluye un plotter 'visualizar_disponibilidad' que usa matplotlib.
    
    Incluye un plotter que cambia el estado de los asientos de las posiciones introducidas 
    
    '''
    
    def __init__(self, posiciones_filas, posiciones_columnas, fila_pantalla, filas_vacias=[], 
                 columnas_vacias=[], posiciones_huecas = []):
        
        # Dimension
        self.n_filas    = len(posiciones_filas)
        self.n_columnas = len(posiciones_columnas)
        # Filas/Columnas
        self.filas_list = posiciones_filas
        self.columnas_list = posiciones_columnas
        # Vacias: Filas/columnas/posiciones
        self.filas_vacias = filas_vacias
        self.columnas_vacias = columnas_vacias
        self.posiciones_huecas = posiciones_huecas
        # Pantalla de cine
        # > posicion centro 
        self.fila_pantalla = fila_pantalla

        
        # Cargamos las posiciones
        posiciones = []
        for fila_pos in posiciones_filas:
            
            # > Nos saltamos las filas vacias
            if fila_pos in filas_vacias: continue
            
            for col_pos in posiciones_columnas:
                
                # > Nos saltamos las columnas vacias
                if col_pos in columnas_vacias: continue
                # > Nos saltamos los asientos huecos
                if (fila_pos, col_pos) in self.posiciones_huecas: 
                    continue
                
                # > Añadimos la posicion del asiento
                posiciones.append((fila_pos, col_pos))
                
        super().__init__(posiciones)
        
        
    # Plotter
    def visualizar_disponibilidad(self, guardar_imagen = False):
        'Usamos matplotlib para mostrar la ocupación de la sala'
        
        plt.figure(figsize=(5,5))
        
        for asiento in self.asientos_dict().values():
            
            if asiento.esta_disponible():
                plt.scatter(asiento.posicion()[1], asiento.posicion()[0], color='green', marker = 's')
            else:
                plt.scatter(asiento.posicion()[1], asiento.posicion()[0], color='red', marker = 's')
        
        
        plt.axhline(y=self.fila_pantalla, xmin=0.2, xmax=0.8, linewidth=6, color='black')
        
        # Quitar ticks
        plt.yticks([])
        plt.xticks([])
        
        # Quitar lineas de recuadro
        for pos in ['right','top','bottom','left']:
            plt.gca().spines[pos].set_visible(False)
            
        # Añadir texto de referencia
        
        col_min = min(self.columnas_list)
        fila_max = max(self.filas_list)
        
        for jj, xx in enumerate(self.columnas_list):
            if xx in self.columnas_vacias: continue
            plt.text(xx, fila_max+1, f'{jj+1}', va='top', ha='center', fontsize=8)
        for ii, yy in enumerate(self.filas_list):
            if yy in self.filas_vacias: continue
            #plt.text(col_min-1, yy, chr(97+ii).upper(),  va='center', ha='right')
            plt.text(col_min-1, yy, f'{ii+1}',  va='center', ha='right', fontsize=8)
        
        if guardar_imagen:
            plt.savefig('estado_sala.png')
        plt.show()
        
        
    
    def ocupar_asientos_dinamico(self, *posiciones_a_ocupar):
        
        for posicion in posiciones_a_ocupar:
            # En vez de comprobar que la posicion del asiento existe
            #  probamos a acceder con ese key
            try:
                # Ocupamos el asiento y mostamos mensaje dependiendo del éxito de la
                #  transacción
                if self.asientos_dict()[posicion].ocupar():
                    print(f'Asiento {posicion} correctamente adquirido.')
                else:
                    print(f'Asiento {posicion} está ocupado.')
                
            except: 
                print(f'Ha habído un error con la posición {posicion}, quizá este asiento no exista.')
                pass
        print('-'*50)

        self.visualizar_disponibilidad()
```

## Varias inicializaciones para casuística

**Añadiremos capacidad para leer desde fichero las posiciones y disponibilidad de los asientos de la sala**

```python
class Sala_cine(Sala):
    ''' 
    Simula una aala de cine mediante instancias de Asiento() en todas las combinaciones de posiciones de fila, **posiciones_filas** \
          y **posiciones_columnas**, excluyendo las posiciones en la variable opcional **huecos**.
          
    Incluye un plotter 'visualizar_disponibilidad' que usa matplotlib.
    
    Incluye un plotter que cambia el estado de los asientos de las posiciones introducidas 
    
    '''
    
    def __init__(self, posiciones_filas, posiciones_columnas, fila_pantalla, filas_vacias=[], 
                 columnas_vacias=[], posiciones_huecas = []):
        
        # Dimension
        self.n_filas    = len(posiciones_filas)
        self.n_columnas = len(posiciones_columnas)
        # Filas/Columnas
        self.filas_list = posiciones_filas
        self.columnas_list = posiciones_columnas
        # Vacias: Filas/columnas/posiciones
        self.filas_vacias = filas_vacias
        self.columnas_vacias = columnas_vacias
        self.posiciones_huecas = posiciones_huecas
        # Pantalla de cine
        # > posicion centro 
        self.fila_pantalla = fila_pantalla

        
        # Cargamos las posiciones
        posiciones = []
        for fila_pos in posiciones_filas:
            
            # > Nos saltamos las filas vacias
            if fila_pos in filas_vacias: continue
            
            for col_pos in posiciones_columnas:
                
                # > Nos saltamos las columnas vacias
                if col_pos in columnas_vacias: continue
                # > Nos saltamos los asientos huecos
                if (fila_pos, col_pos) in self.posiciones_huecas: continue
                
                # > Añadimos la posicion del asiento
                posiciones.append((fila_pos, col_pos))
                
        super().__init__(posiciones)
    
        # Obtener las posiciones de filas/columnas que tienen asientos
        filas_list = []
        columnas_list = []
        for pos in posiciones:
            filas_list.append(pos[0])
            columnas_list.append(pos[1])
        
    ######################## CÓDIGO AÑADIDO ################################################################
    @classmethod # Decorador
    def desde_fichero(cls, path_completo):
        
        # Leemos fichero
        sala_cine_pd = pd.read_csv(path_completo)
        
        # ! Necesitamos asegurarnos que las columnas del fichero sean (fila_pos, colum_pos, disponibilidad)
        
        posiciones_filas = sala_cine_pd['fila_pos'].unique()
        posiciones_columnas = sala_cine_pd['colum_pos'].unique()
        
        posiciones_disponibles = []
        posiciones_ocupadas    = []
        for fila_fichero in range(sala_cine_pd.shape[0]):
            if sala_cine_pd.iloc[fila_fichero, 2].lower() == 'disponible':
                posiciones_disponibles.append( tuple(sala_cine_pd.iloc[fila_fichero,0:2].to_list() ))
            else:
                posiciones_ocupadas.append(  tuple(sala_cine_pd.iloc[fila_fichero,0:2].to_list() ))
        
        
        # Todas las posiciones posibles
        
        huecos = []
        
        for fila in posiciones_filas:
            for col in posiciones_columnas:
                if (fila, col) not in (set(posiciones_disponibles) | set(posiciones_ocupadas)):
                    huecos.append( (fila, col) )
        
        # Llamamos al constructor y ocupamos asientos no disponibles
        aux = cls(posiciones_filas, posiciones_columnas, -3, posiciones_huecas = huecos)
        aux.ocupar_asientos(*posiciones_ocupadas)
        aux.cargado_de_archivo(path_completo)
        return aux
    
    @staticmethod
    def cargado_de_archivo(path):
        print(f'Sala de cine cargada de {path}')
        
    
    #########################################################################################################
    
        
    # Plotter
    def visualizar_disponibilidad(self):
        'Usamos matplotlib para mostrar la ocupación de la sala'
        

        plt.figure(figsize=(5,5))
        
        
        for asiento in self.asientos_dict().values():
            
            if asiento.esta_disponible():
                plt.scatter(asiento.posicion()[1], asiento.posicion()[0], color='green', marker = 's')
            else:
                plt.scatter(asiento.posicion()[1], asiento.posicion()[0], color='red', marker = 's')
        
        
        plt.axhline(y=self.fila_pantalla, xmin=0.2, xmax=0.8, linewidth=6, color='black')
        
        # Quitar ticks
        plt.yticks([])
        plt.xticks([])
        
        # Quitar lineas de recuadro
        for pos in ['right','top','bottom','left']:
            plt.gca().spines[pos].set_visible(False)
            
        # Añadir texto de referencia
        
        col_min = min(self.columnas_list)
        fila_max = max(self.filas_list)
        
        for jj, xx in enumerate(self.columnas_list):
            if xx in self.columnas_vacias: continue
            plt.text(xx,fila_max+1, f'{jj+1}', va='top', ha='center', fontsize=8)
        for ii, yy in enumerate(self.filas_list):
            if yy in self.filas_vacias: continue
            #plt.text(col_min-1, yy, chr(97+ii).upper(),  va='center', ha='right')
            plt.text(col_min-1, yy, ii+1,  va='center', ha='right', fontsize=8)
            
            
        
        plt.show()
                     
    
    def ocupar_asientos_dinamico(self, *posiciones_a_ocupar):
        
        for posicion in posiciones_a_ocupar:
            # En vez de comprobar que la posicion del asiento existe
            #  probamos a acceder con ese key
            try:
                # Ocupamos el asiento y mostamos mensaje dependiendo del éxito de la
                #  transacción
                if self.asientos_dict()[posicion].ocupar():
                    print(f'Asiento {posicion} correctamente adquirido.')
                else:
                    print(f'Asiento {posicion} está ocupado.')
                
            except:
                print(f'Ha habído un error con la posición {posicion}, quizá este asiento no exista.')
                pass
        print('-'*50)

        self.visualizar_disponibilidad()
        
```
